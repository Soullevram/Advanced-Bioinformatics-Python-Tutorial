```python
# Question 1
## Write a function that will parse fasta files with biopython.

# Import SeqIO module
from Bio import SeqIO
def parse_fasta(file_path):
    """
    Parses a FASTA file using SeqIO module and returns a dictionary with sequence IDs as keys and sequences as values.
    
    Parameter 
    file_path(str): Path to the FASTA file.
    
    Return
    Dictionary {sequence_id: sequence}
    """
    # Create an empty dictionary
    sequences = {}
    # Open file
    with open(file_path, "r") as fasta_file:
        # For loop to iterate through lines
        for record in SeqIO.parse(fasta_file, "fasta"):
            # Extract the sequence ID and store it in the dictiontary
            sequences[record.id] = str(record.seq)
    return sequences

# Example usage:
fasta_sequences = parse_fasta("Downloads/AB_test_data.fasta")
print(fasta_sequences)
```

```python
# Question 3
## Write a piece of code that will run a grep command through python and will deal with the output meaningfully.
import subprocess

def grep_search(pattern, file_path):
    try:
        # Run the grep command and capture output
        result = subprocess.run(
            ['grep', pattern, file_path], 
            text=True, capture_output=True, check=True
        )
        
        # Process the output
        matched_lines = result.stdout.strip().split('\n')
        return matched_lines

    except subprocess.CalledProcessError as e:
        # grep exits with code 1 if no matches are found
        if e.returncode == 1:
            return []  # No matches found
        else:
            raise  # Re-raise error for other issues

# Example usage
file_path = "Downloads/AB_test_data.fasta"
pattern = "pop"
matches = grep_search(pattern, file_path)

if matches:
    print(f"Found {len(matches)} matches:")
    for line in matches:
        print(line)
else:
    print("No matches found.")
```

```python
# Question 4
## Write some python code to create a new directory, and then change your working directory to the new directory.

# Import OS module
import os

# Define the new directory name
new_directory = "my_new_directory"

# Create the directory if it doesn't exist
if not os.path.exists(new_directory):
    os.makedirs(new_directory)

# Change the working directory to the new directory
os.chdir(new_directory)

# Print the current working directory to verify
print("Current Working Directory:", os.getcwd())
```
