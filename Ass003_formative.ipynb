```python
#3
# import OS module
import os

# Define file path, which is, for me, in my Downloads
downloads_path = os.path.expanduser("~/Downloads")
file_name = "kraken_taxonomic.kreport"  
file_path = os.path.join(downloads_path, file_name)

# Create an empty set to avoid duplicates and ensure uniqueness
unique_species = set()

# Read file and process lines
with open(file_path, "r", encoding="utf-8") as file:
    # For loop to iterate through the lines
    for line in file:
        # to split the tab-separated values
        columns = line.strip().split("\t")  
        # Check if 4th column has a 'S', which means species
        if len(columns) > 5 and columns[3] == "S":  
            # Extract species name from the 6th column; the number 5 is used because python starts indexing/counting from 0
            species_name = columns[5].strip()  
            # Add species to set if unique
            unique_species.add(species_name)  

# Print unique species
for species in sorted(unique_species):
    print(species)
```

```python
#4
def gc_content(sequence):
    """Calculate the GC content percentage of a DNA sequence.
    Parameter:
    sequence (str): A DNA sequence consisting of 'A', 'T', 'G', and 'C'.

    Returns:
    float: The percentage of the sequence that is composed of 'G' and 'C'.
    """
    gc_count = sequence.count('G') + sequence.count('C')
    return (gc_count / len(sequence)) * 100

# Assign the sequences to a list variable
Seq_list = [
    "ATGCGCGTGCGTACGTGCTAAACGTGCATGC",
    "ATGGGCGTCAGTCTGACTGAAATCTTGCGTACT",
    "ATGCCGTCTACGTAACCTGCTAAAAACTGCA",
    "ATGCGTATGCACACACTGTCAATGCAT",
    "ATGCTGCAAATGCCATCACTGCAATCGAA"
]

# Filter sequences with GC content > 55%
filtered_sequences = [seq for seq in Seq_list if gc_content(seq) > 55]

# Print the new list containing sequence(s) with >55% GC content
print(filtered_sequences)
```

```python
#5
def convert_fasta_format(input_file, output_file):
    """
    Convert a Type-two FASTA file into a Type-one format.

    Parameters:
    input_file (str): Path to the input FASTA file (multi-line format).
    output_file (str): Path to the output FASTA file (single-line format).
    
    Returns:
    None: Writes the reformatted sequences to the output file.
    """
    with open(input_file, "r") as infile, open(output_file, "w") as outfile:
        # Create an empty list
        sequence_lines = []
        # For loop to iterate through the lines
        for line in infile:
            line = line.strip()
            # If it's a header line
            if line.startswith(">"):  
                if sequence_lines:
                    # Write previous sequence
                    outfile.write("".join(sequence_lines) + "\n")  
                    sequence_lines = []
                    # Write header
                outfile.write(line + "\n")  
            else:
                # Collect sequence parts
                sequence_lines.append(line)  
        
        if sequence_lines:
            # Write last sequence
            outfile.write("".join(sequence_lines) + "\n")  

# Example 
example_fasta = "Downloads/Sequence_1.txt"   
output_fasta = "output.fasta" 
convert_fasta_format(example_fasta, output_fasta)

print(f"Conversion complete. Output saved to {output_fasta}")
# To view the new fasta format
with open("output.fasta", "r") as file:
    print(file.read()) 
```

```python
#7
def count_lines(file_path):
    """
    Counts the number of lines in a text file.

    Parameters:
    file_path (str): Path to the text file.

    Returns:
    int: The total number of lines in the file.
    """
    with open(file_path, "r") as file:
        return sum(1 for _ in file)

# Example
file_path = "Downloads/Sequence_1.txt"  # File path from question 5
num_lines = count_lines(file_path)
print(f"Total lines in the file: {num_lines}")
```

```python
#8
def count_sequence_differences(seq_1, seq_2):
    """
    Counts the number of differences (mismatches) between two sequences of equal length.

    Parameters:
    seq_1 (str): The first sequence.
    seq_2 (str): The second sequence.

    Returns:
    int: The number of positions where the sequences differ.

    Raises:
    ValueError: If the sequences are not of equal length.
    """
    # If statement to ensure that sequence strings are of the same length
    if len(seq_1) != len(seq_2):
        raise ValueError("Sequences must be of equal length.")

    return sum(  
    1  # Adds 1 for every mismatch found  
    for a, b in zip(seq_1, seq_2)  # Pairs up corresponding characters/positions from both sequences  
    if a != b  # Counts only when characters at the same position are different  
)

# Example 
seq_1 = "TTTCTATA"
seq_2 = "TATATATA"

differences = count_sequence_differences(seq_1, seq_2)
print(f"Number of differences: {differences}")
```

```python
#9
def aln_2_fasta(file):
    """
    Converts a CLUSTAL alignment file into a dictionary format suitable for FASTA conversion.

    Parameters:
    file (str): The path to the CLUSTAL alignment file.

    Returns:
    dict: A dictionary where keys are sequence headers and values are lists of sequence fragments.
    """
    # Import the regular expressions module for pattern matching
    import re  
    # Create an empty dictionary to store sequence headers and sequences
    aln_dict = {}  

    # Open the clustal alignment file in read mode
    with open(file, "r") as tmp_in:
        # For loop to iterate over each line in the file
        for lines in tmp_in:  
            # Ignore header lines that start with "CLUSTAL"
            if lines.startswith("CLUSTAL"):
                # Do nothing and continue to the next line
                pass  

            # Process lines that are not empty and do not contain "*"
            elif not lines.startswith("\n") and not "*" in lines:
                # Use regex to extract non-whitespace elements from the line
                lines = re.findall(r'\S+', lines)

                # Ensure the extracted line contains data
                if len(lines) > 0:
                    # To extract sequence header and sequence part
                    header, seq_parts = lines[0], lines[1]  

                    # If the header already exists in the dictionary, append the sequence part
                    if header in aln_dict:
                        aln_dict[header].append(seq_parts)
                    else:
                        # If not, create a new entry in the dictionary
                        aln_dict[header] = [seq_parts]  

    # Return the dictionary containing sequence headers and sequences
    return aln_dict  
clustal_file = "Downloads/clustal_aln.txt"
clustal_2 = aln_2_fasta(clustal_file)
print(clustal_2)
```

```python
#10
def is_bigger_than(mini, maxi, threshold):
    """
    Generates a list of numbers within a given range where each number,
    when multiplied by 27, exceeds a specific threshold.

    Parameters:
    mini (int): The starting value of the range (inclusive).
    maxi (int): The stopping value of the range (exclusive).
    threshold (int): The value that "x * 27" must exceed.

    Returns:
    list: A list of numbers that meet the condition.
    """
    # Create an empty list to store the numbers meet the requirement
    numbers = []  

    # For loop to iterate through numbers starting from 'mini' to 'maxi', increasing by 5 each time
    for x in range(mini, maxi, 5):
        # Check if the number multiplied by 27 is greater than the threshold
        if x * 27 > threshold:
            # If true, append the number to the list
            numbers.append(x)  
    # Return a list containing numbers that satisfy the requirement
    return numbers  

# Example:
test_example = is_bigger_than(0, 100, 500)  # Finds numbers in range(0, 100, 5) where x * 27 > 500
print(test_example)  
```
